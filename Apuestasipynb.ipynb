```python
import numpy as np

def create_betting_target(df, prob_home_col='prob_home_win', prob_away_col='prob_away_win', threshold=0.5):
    """
    Crea la columna target 'apostar / no apostar' basada en un umbral de probabilidad.

    Args:
        df (DataFrame): dataset que contiene las columnas de probabilidades
        prob_home_col (str): nombre de la columna con probabilidad de victoria local
        prob_away_col (str): nombre de la columna con probabilidad de victoria visitante
        threshold (float): probabilidad mínima para considerar apostar

    Returns:
        DataFrame: dataset con columna 'target' agregada
    """
    df = df.copy()
    df['target'] = np.where(
        (df[prob_home_col] > threshold) | (df[prob_away_col] > threshold),
        1,
        0
    )
    return df
```

```python
# --- Escalar algunas variables para que estén en rango comparable ---
df['home_ppda_scaled'] = df['home_ppda'] / (df['home_ppda'] + df['away_ppda'])
df['away_ppda_scaled'] = df['away_ppda'] / (df['home_ppda'] + df['away_ppda'])

df['home_deep_scaled'] = df['home_deep_completions'] / (df['home_deep_completions'] + df['away_deep_completions'])
df['away_deep_scaled'] = df['away_deep_completions'] / (df['home_deep_completions'] + df['away_deep_completions'])

# Diferencia de NP-xG
df['home_np_diff_scaled'] = df['home_np_xg_difference'] / (abs(df['home_np_xg_difference']) + abs(df['away_np_xg_difference']))
df['away_np_diff_scaled'] = df['away_np_xg_difference'] / (abs(df['home_np_xg_difference']) + abs(df['away_np_xg_difference']))
```

```python
# Ventaja local
home_adv = 0.05

# --- Ponderaciones ajustables ---
w_xg = 0.5
w_ppda = 0.2
w_deep = 0.2
w_np = 0.1
draw_factor = 0.5

# --- Cálculo de probabilidades ---
df['prob_home_win'] = (
    w_xg*(df['home_xg'] / (df['home_xg'] + df['away_xg'])) +
    w_ppda*df['home_ppda_scaled'] +
    w_deep*df['home_deep_scaled'] +
    w_np*df['home_np_diff_scaled'] +
    home_adv
)

df['prob_away_win'] = (
    w_xg*(df['away_xg'] / (df['home_xg'] + df['away_xg'])) +
    w_ppda*df['away_ppda_scaled'] +
    w_deep*df['away_deep_scaled'] +
    w_np*df['away_np_diff_scaled'] -
    home_adv
)

# Probabilidad de empate: lo que "queda" para que todo sume 1
df['draw_raw'] = np.exp(-np.abs(df['home_xg'] - df['away_xg']))

# Normalizar para que sumen 1 con el empate
total = df['prob_home_win'] + df['prob_away_win'] + df['draw_raw']
df['prob_home_win'] /= total
df['prob_away_win'] /= total
df['prob_draw'] /= total
```

```python
# --- Crear target apostar / no apostar ---
df = create_betting_target(df, prob_home_col='prob_home_win', prob_away_col='prob_away_win', threshold=0.5)

# --- Revisar resultados ---
df[['home_team','away_team','prob_home_win','prob_away_win','prob_draw','target']].head(10)
```

```python
#Forma Reciente del equipo
import pandas as pd
import numpy as np

# Definimos ventana (número de partidos a considerar)
window = 5

# Lista de features que queremos hacer rolling
rolling_cols = ['home_xg', 'away_xg', 'home_np_xg', 'away_np_xg', 'home_ppda', 'away_ppda', 'home_deep_completions', 'away_deep_completions']

# Creamos un dataframe vacío para guardar todos los rolling features
rolling_features = pd.DataFrame()

# Obtenemos lista de equipos
teams = pd.concat([df['home_team'], df['away_team']]).unique()

for team in teams:
    # Filtramos partidos donde participó el equipo
    team_df = df[(df['home_team'] == team) | (df['away_team'] == team)].sort_values('date')

    # Creamos dataframe temporal
    temp = pd.DataFrame()
    temp['game_id'] = team_df['game_id']

    # Rolling features por columna
    for col in rolling_cols:
        temp[f'{col}_rolling_mean'] = team_df[col].rolling(window=window, min_periods=1).mean().values

    temp['team'] = team
    rolling_features = pd.concat([rolling_features, temp], axis=0)

# Revisamos resultados
rolling_features.head()
```

```python
def resumen_apuesta_rentable(df, rolling_df, partido_idx, cuotas, n_last=5):
    """
    Genera un resumen de apuesta rentable de un partido con rolling features, EV y últimos rivales.
    """
    partido = df.iloc[partido_idx]
    home_team = partido['home_team']
    away_team = partido['away_team']
    game_id = partido['game_id']

    # Rolling del equipo local
    home_rolling = rolling_df[rolling_df['team']==home_team].set_index('game_id').loc[game_id]
    # Rolling del equipo visitante
    away_rolling = rolling_df[rolling_df['team']==away_team].set_index('game_id').loc[game_id]

    # Probabilidades implícitas
    p_home = partido['prob_home_win']
    p_away = partido['prob_away_win']
    p_draw = partido['prob_draw']

    # EV (Valor esperado)
    ev_home = p_home * cuotas['home'] - 1
    ev_away = p_away * cuotas['away'] - 1
    ev_draw = p_draw * cuotas['draw'] - 1

    # -------------------------------
    # ÚLTIMOS PARTIDOS DEL LOCAL
    # -------------------------------
    home_history = df[
        (df['home_team']==home_team) | (df['away_team']==home_team)
    ].sort_values('date').tail(n_last)

    home_last = []
    for _, row in home_history.iterrows():
        if row['home_team'] == home_team:
            rival = row['away_team']
            xg_for = row['home_xg']
            xg_against = row['away_xg']
            result = "W" if row['home_goals'] > row['away_goals'] else ("L" if row['home_goals'] < row['away_goals'] else "D")
            venue = "Local"
        else:
            rival = row['home_team']
            xg_for = row['away_xg']
            xg_against = row['home_xg']
            result = "W" if row['away_goals'] > row['home_goals'] else ("L" if row['away_goals'] < row['home_goals'] else "D")
            venue = "Visitante"

        home_last.append(f"{venue} vs {rival}: {result} (xG {xg_for:.2f}-{xg_against:.2f})")

    # -------------------------------
    # ÚLTIMOS PARTIDOS DEL VISITANTE
    # -------------------------------
    away_history = df[
        (df['home_team']==away_team) | (df['away_team']==away_team)
    ].sort_values('date').tail(n_last)

    away_last = []
    for _, row in away_history.iterrows():
        if row['home_team'] == away_team:
            rival = row['away_team']
            xg_for = row['home_xg']
            xg_against = row['away_xg']
            result = "W" if row['home_goals'] > row['away_goals'] else ("L" if row['home_goals'] < row['away_goals'] else "D")
            venue = "Local"
        else:
            rival = row['home_team']
            xg_for = row['away_xg']
            xg_against = row['home_xg']
            result = "W" if row['away_goals'] > row['home_goals'] else ("L" if row['away_goals'] < row['home_goals'] else "D")
            venue = "Visitante"

        away_last.append(f"{venue} vs {rival}: {result} (xG {xg_for:.2f}-{xg_against:.2f})")

    # -------------------------------
    # RESUMEN FINAL
    # -------------------------------
    resumen = f"""
Partido: {home_team} vs {away_team} ({partido['date'].date()})

Forma reciente del equipo local ({home_team}):
- xG promedio últimos 5 partidos: {home_rolling['home_xg_rolling_mean']:.2f}
- NP-xG promedio: {home_rolling['home_np_xg_rolling_mean']:.2f}
- PPDA: {home_rolling['home_ppda_rolling_mean']:.2f}
- Deep completions: {home_rolling['home_deep_completions_rolling_mean']:.2f}

Últimos {n_last} rivales del {home_team}:
{chr(10).join(home_last)}

--------------------------------------------

Forma reciente del equipo visitante ({away_team}):
- xG promedio últimos 5 partidos: {away_rolling['away_xg_rolling_mean']:.2f}
- NP-xG promedio: {away_rolling['away_np_xg_rolling_mean']:.2f}
- PPDA: {away_rolling['away_ppda_rolling_mean']:.2f}
- Deep completions: {away_rolling['away_deep_completions_rolling_mean']:.2f}

Últimos {n_last} rivales del {away_team}:
{chr(10).join(away_last)}

--------------------------------------------

Probabilidades implícitas:
- Victoria local: {p_home:.2f}
- Empate: {p_draw:.2f}
- Victoria visitante: {p_away:.2f}

Cuotas de la casa:
- Local: {cuotas['home']}
- Empate: {cuotas['draw']}
- Visitante: {cuotas['away']}

Valor esperado (EV):
- EV local: {ev_home:.2f}
- EV empate: {ev_draw:.2f}
- EV visitante: {ev_away:.2f}

Apuesta recomendada: {'Sí' if partido['target']==1 else 'No'}
"""
    return resumen
```
